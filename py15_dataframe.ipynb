# 시간에 따른 TIp의 차이?

* 시간 별 Tip의 평균

```python
x = tips.groupby('sex')['tip'].mean()
x1 = pd.DataFrame(x)
```

```python
x1
```

```python
y = tips.groupby('sex')['tip'].min()
y1 = pd.DataFrame(y)
```

```python
y1
```

```python
z = tips.groupby('sex')['tip'].max()
z1 = pd.DataFrame(z)
```

```python
z1
```

```python
x2 = tips.groupby(['smoker','sex'])['tip'].mean()
```

```python
pd.DataFrame(x2)
```

```python
x3 = tips.groupby(['day','sex'])['tip'].mean()
pd.DataFrame(x3)
```

```python
x4 = tips.groupby(['time','sex'])['tip'].mean()
pd.DataFrame(x4)
```

```python
x5 = tips.groupby(['sex','day','time'])['tip'].agg({'mean','max','min'})
x5
```
