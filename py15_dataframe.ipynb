## 두 개 이상이 카테고리 별 GroupBy

```python
# 성별, 흡연여부별 Tip의 평균
```

```python
# 성별 요일별 tip의 평균

tips.groupby(['sex','day']).tip.mean()

for x in tips.sex.unique():

  for y in tips.day.unique():



    x6 = tips[(tips.sex==x) & (tips.day==y)]['tip'].mean()

    print(f'{x} / {y} = {x6}')
```

```python
# 성별 영수증, 팁의 평균.

tips.groupby('sex')['tip','total_bill'].agg({'mean','mean'})

for x in tips['sex'].unique():

  y=tips[tips['sex']==x][['total_bill','tip']].mean()

  print(f'{x} \n {y}')
```

```python
# 성별, 흡연 여부 별 영수증, 팁의 평균

tips.groupby(['sex','smoker'])[['tip','total_bill']].mean()

for x in tips.sex.unique():

  for y in tips.smoker.unique():

    z = tips[(tips.sex == x) & (tips.smoker == y)][['tip','total_bill']].mean()
    print(f'{x} / {y} \n {z}')
```

## 그룹별 2개 이상의 통계 함수 적용

```python
# 성별 Tip의 평균, 최댓값, 중앙값, 최솟값



tips.groupby('sex')['tip'].agg({'mean','min','median','max'})
```

```python
# 성별 흡연 여부별 영수증과 팁의 평균과 표준편차

y = tips.groupby(['sex','smoker'])[['tip','total_bill']].agg(['mean','std'])
pd.DataFrame(y)
```

# `pivot_table` 함수(메서드)

* Parameters :
  * values : 통계 함수를 적용할 변수(컬럼)
  * index : 결과 데이터프레임에서 행 인덱스로 사용할 변수(컬럼)
  * columns : 결과 데이터프레임에서 열 인덱스로 사용할 변수(컬럼)
  * aggfunc : aggregation function. 기본값은 평균(mean).

```python
# 성별 팁의 평균.

tips.pivot_table(values='tip', index='sex') # aggfunc = 'mean' (기본값)
```

```python
# 성별 흡연여부별 팁의 평균

tips.pivot_table(values='tip', index=['sex','smoker'])

for x in tips.sex.unique():

  for y in tips.smoker.unique():

    z = tips[(tips['sex']==x) & (tips['smoker']==y)]['tip'].mean()
    print(f'{x} / smoking = {y} \ntip_mean = {z}')
```

```python
tips.pivot_table(values='tip', index='sex', columns='smoker')
```

```python
# 성별, 흡연여부별, 요일별, 시간대별, 팁의 평균

tips.pivot_table(fill_value=0, values='tip', index=['sex','smoker'], columns=['day','time'])
```

```python
# 요일별, 시간별 영수증과 팁의 평균, 표준편차

tips.pivot_table(values=['total_bill','tip'], index='day', columns='time', aggfunc=['mean','std'], fill_value='ValueError')
```
