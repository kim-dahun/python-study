```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mpg = sns.load_dataset('mpg')
```

```python
mpg.head()
```

```python
mpg.info()
```

mpg(Mile Per Gallon) dataset: 자동차 연비 데이터 셋

* mpg: 연비
* cylinders: 실린더 개수
* displacement: 배기량
* horsepower: 마력
* weight: 자동차 무게
* acceleration: 가속도.
* model_year: 자동차 생산 연도
* origin: 자동차 생산지
* name: 자동차 이름

* mpg 데이터 셋 탐색
* mpg ~ displacement scatter plot
* 2x2 subplot 을 생성. scatterplot

  * mpg ~ cylinders
  * mpg ~ horsepower
  * mpg ~ weight
  * mpg ~ acceleration

* cylinders 별 mpg 평균 막대 그래프
* cylinders 별 mpg boxplot
* origin별, cyilinder 별 mpg box plot
* origin별, cyilinder 별 mpg 평균 막대 그래프

```python
mpg.describe()
```

```python
sns.scatterplot(data=mpg, x='mpg',y='displacement')
```

```python
fig, ax = plt.subplots(ncols=2, nrows=2, figsize=(12,10))


ax[0,0].scatter(x=mpg.mpg, y=mpg.cylinders)
ax[0,1].scatter(x=mpg.mpg, y=mpg.horsepower)
ax[1,0].scatter(x=mpg.mpg, y=mpg.weight)
ax[1,1].scatter(x=mpg.mpg, y=mpg.acceleration)

```

```python
sns.barplot(data=mpg, x='cylinders', y='mpg')
```

```python
sns.boxplot(data=mpg, x='cylinders', y='mpg')
```

```python


sns.boxplot(data=mpg, x='origin', y='mpg', hue='cylinders', width=0.6)
```

```python
sns.barplot(data=mpg, x='origin', hue='cylinders', y='mpg')
```
